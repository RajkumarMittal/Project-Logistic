```python
import numpy as np
import pandas as pd
# import pandas as pd
df=pd.read_csv('original_train_titanic.csv')
```

```python
df
```

```python
del df["Ticket"]
```

```python
df.describe()
```

```python
del df["Name"]
```

```python
def h(n):
    a = np.round(n)
    return(int(a))
    pass
df["Fare"] = df.Fare.apply(h)
```

```python
df
```

```python
def h(s):
    if s == "S":
        return 1
    elif s == "Q":
        return 2
    else:
        return 3
    pass
df["Embarked"] = df.Embarked.apply(h)
```

```python
df.describe()
```

```python
def g(s):
    if s == "male":
        return 0
    else:
        return 1
    pass
df["Sex"] = df.Sex.apply(g)
```

```python
a = df[df.Pclass == 1]
sur_1_age = a.Age.mean()
sur_1_age = np.round(sur_1_age)
print(sur_1_age)

b = df[df.Pclass == 2]
sur_2_age = b.Age.mean()
sur_2_age = np.round(sur_2_age)
print(sur_2_age)

c = df[df.Pclass == 3]
sur_3_age = c.Age.mean()
sur_3_age = np.round(sur_3_age)
print(sur_3_age)
```

```python
def f(n):
    Age = n[0]
    P = n[1]
    if pd.isnull(Age):
        if P == 1:
            return int(sur_1_age)
        elif P == 2:
            return int(sur_2_age)
        else:
            return int(sur_3_age)
    else:
        return int(Age)
        
df["Age"] = df[["Age", "Pclass"]].apply(f,axis = 1)
```

```python
df["Age"]
```

```python
df.describe()
```

```python
del df["Cabin"]
```

```python
data = df.to_numpy()
```

```python
x = data[:,:-1]
# import sklearn.preprocessing as pp
# x = pp.scale(x)
y = data[:,-1]
print(x.shape)
print(y.shape)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(C=2,multi_class='auto')
clf.fit(x,y)
```

```python
clf.score(x,y)
```

```python
clf.predict_proba(x)[1]
```

```python
clf.predict(x) - y
```
